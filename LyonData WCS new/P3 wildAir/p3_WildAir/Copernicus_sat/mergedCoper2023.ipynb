* Merging les données de vent et temperature (insta) avec l'accum (precipitation et neige) pour l'année de 2023

```python
import pandas as pd

dataCop_accum23 = pd.read_csv(r"C:\Users\sophi\FrMarques\LyonData WCS new\P3 wildAir\p3_WildAir\Copernicus_sat\dfcopaccum_lyon23.csv")
dataCop_insta23 = pd.read_csv(r"C:\Users\sophi\FrMarques\LyonData WCS new\P3 wildAir\p3_WildAir\Copernicus_sat\dfcopinst_lyon23.csv")
```
