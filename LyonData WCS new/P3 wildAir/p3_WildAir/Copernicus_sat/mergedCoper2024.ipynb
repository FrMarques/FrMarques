```python
def kelvin_to_celsius(df, column):
    """
    Transforms a column with Kelvin temperatures into Celsius.

    Parameters:
        df (pd.DataFrame): The dataframe containing the data.
        column (str): The name of the column with values in Kelvin (float).
    Returns:
        pd.DataFrame: The updated dataframe with the converted values.
    """
    df[column] = df[column] - 273.15
    return df
```

```python
import numpy as np

# Adicionar as colunas de velocidade e direção do vento
def calculate_wind_speed_direction(df):
    """
    Calcula a velocidade e direção do vento a partir de u10 e v10.

    Parâmetros:
        df (pd.DataFrame): O dataframe que contém as colunas u10 e v10.

    Retorna:
        pd.DataFrame: O dataframe atualizado com as colunas 'wind_speed' e 'wind_direction'.
    """
    # Velocidade do vento
    df["wind_speed"] = np.sqrt(df["u10"]**2 + df["v10"]**2)

    # Direção do vento em graus
    df["wind_direction"] = np.degrees(np.arctan2(-df["u10"], -df["v10"]))
    df["wind_direction"] = df["wind_direction"] % 360  # Ajustar para o intervalo [0, 360]

    return df

# Aplicar ao dataset
merged_df = calculate_wind_speed_direction(merged_df)

# Visualizar as novas colunas
print(merged_df[["u10", "v10", "wind_speed", "wind_direction"]].head())
```
