```python
def kelvin_to_celsius(df, column):
    """
    Transforms a column with Kelvin temperatures into Celsius.

    Parameters:
        df (pd.DataFrame): The dataframe containing the data.
        column (str): The name of the column with values in Kelvin (float).
    Returns:
        pd.DataFrame: The updated dataframe with the converted values rounded to just one decimal case.
    """
    df[column] = (df[column] - 273.15).round(1)
    return df
```

```python
dataCop_lyon2324_C = dataCop_lyon2324.copy()
dataCop_lyon2324_C = kelvin_to_celsius(dataCop_lyon2324_C, "t2m")
```

```python
import numpy as np

  # Adding columns for the speed and direction of the wind:
def calculate_wind_speed_direction(df):
    """
    Calculates speed and direction of the wind from "u10" and "v10" values.

    Parameters:
        df (pd.DataFrame): The dataframe with the columns "u10" and "v10".

    Returns:
        pd.DataFrame: The same dataframe updated with the new columns: 'wind_speed' e 'wind_direction'.
    """
      # Wind speed
    df["wind_speed"] = np.sqrt(df["u10"]**2 + df["v10"]**2)

      # Wind direction in degrees:
    df["wind_direction"] = np.degrees(np.arctan2(-df["u10"], -df["v10"]))
    df["wind_direction"] = df["wind_direction"] % 360  # Adjusting to the range [0, 360]:

    return df
```

```python
  # Applying to the dataset
dataCop_lyon2324_C_W = dataCop_lyon2324_C.copy()
dataCop_lyon2324_C_W = calculate_wind_speed_direction(dataCop_lyon2324_C_W)
```

```python
# Function to convert wind direction angles into text direction labels:
def wind_direction_label(angle):
    if 337.5 <= angle <= 360 or 0 <= angle < 22.5:
        return "N"
    elif 22.5 <= angle < 67.5:
        return "NE"
    elif 67.5 <= angle < 112.5:
        return "E"
    elif 112.5 <= angle < 157.5:
        return "SE"
    elif 157.5 <= angle < 202.5:
        return "S"
    elif 202.5 <= angle < 247.5:
        return "SO"
    elif 247.5 <= angle < 292.5:
        return "O"
    elif 292.5 <= angle < 337.5:
        return "NO"
    else:
        return None  # In case of a not expected error if values higher than 360
```

```python
    # Generating a new dataframe to include the wind direction labels column:
dataCop_lyon2324_C_Wd = dataCop_lyon2324_C_W.copy()
dataCop_lyon2324_C_Wd["wind_direction_label"] = dataCop_lyon2324_C_Wd["wind_direction"].apply(wind_direction_label)
```

```python
dataCop_lyon2324_C_Wd["wind_speed"] = dataCop_lyon2324_C_Wd["wind_speed"].round(2)
```
